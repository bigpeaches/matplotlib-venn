## Venn3 :: Visual review of special cases

```python
from matplotlib_venn import venn3
```

```python
venn3((1, 1, 1, 1, 1, 1, 1));
```

```python
# |ABC| = 0
venn3((1, 1, 1, 1, 1, 1, 0));
```

```python
# |ABc| = 0 / |aBC| = 0 / |AbC| = 0 / |ABC| != 0
figure(figsize=(9,3))
subplot(131)
venn3((1, 1, 0, 1, 1, 1, 1));
subplot(132)
venn3((1, 1, 1, 1, 0, 1, 1));
subplot(133)
venn3((1, 1, 1, 1, 1, 0, 1));
```

```python
# |ABc| = 0 / |aBC| = 0 / |AbC| = 0 / |ABC| == 0
figure(figsize=(9,3))
subplot(131)
venn3((1, 1, 0, 1, 1, 1, 0));
subplot(132)
venn3((1, 1, 1, 1, 0, 1, 0));
subplot(133)
venn3((1, 1, 1, 1, 1, 0, 0));
```

```python
# |Abc| = 0 / |aBc| = 0 / |abC| = 0 / |ABC| != 0
figure(figsize=(9,3))
subplot(131)
venn3((0, 1, 1, 1, 1, 1, 1));
subplot(132)
venn3((1, 0, 1, 1, 1, 1, 1));
subplot(133)
venn3((1, 1, 1, 0, 1, 1, 1));
```

```python
# |Abc| = 0 / |aBc| = 0 / |abC| = 0 / |ABC| == 0
figure(figsize=(9,3))
subplot(131)
venn3((0, 1, 1, 1, 1, 1, 0));
subplot(132)
venn3((1, 0, 1, 1, 1, 1, 0));
subplot(133)
venn3((1, 1, 1, 0, 1, 1, 0));
```

```python
# Two nonintersecting sets
figure(figsize=(9,3))
subplot(131)
venn3([set([1,2]), set([2,3]), set([3,4])]);
subplot(132)
venn3([set([2,3]), set([3,4]), set([1,2])]);
subplot(133)
venn3([set([3,4]), set([1,2]), set([2,3])]);
```

```python
# All sets nonintersecting
venn3([set([1]), set([2]), set([3])]);
```

```python
# One set subset of another, intersecting with third
subplot(131)
venn3([set([1,2]), set([1,2,3]), set([2,3])]);
subplot(132)
venn3([set([1,2,3]), set([2,3]), set([1,2])]);
subplot(133)
venn3([set([2,3]), set([1,2]), set([1,2,3])]);
```
